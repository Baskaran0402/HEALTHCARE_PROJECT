```python
import numpy as np

# Convert symptoms into binary vector
def symptoms_to_vector(symptoms_list, all_symptoms):
    return [[1 if symptom in symptoms_list else 0 for symptom in all_symptoms]]

# Predict disease from symptoms
def predict_disease(symptoms_list):
    X_input = symptoms_to_vector(symptoms_list, all_symptoms)
    y_pred = disease_model.predict(X_input)
    predicted_disease = label_encoder.inverse_transform(y_pred)[0]
    return predicted_disease

# Get risk score from a binary classifier
def get_risk_score(model, X_input):
    if hasattr(model, "predict_proba"):
        prob = model.predict_proba(X_input)[0][1]  # probability of positive class
        return prob * 100
    else:
        return model.predict(X_input)[0] * 100  # fallback if no predict_proba

# Unified Health Risk Engine
def health_risk_engine(X_input, symptoms_list):
    scores = {}
    scores["heart"] = get_risk_score(loaded_models["heart"], X_input)
    scores["diabetes"] = get_risk_score(loaded_models["diabetes"], X_input)
    scores["kidney"] = get_risk_score(loaded_models["kidney"], X_input)
    scores["liver"] = get_risk_score(loaded_models["liver"], X_input)
    scores["stroke"] = get_risk_score(loaded_models["stroke"], X_input)
    scores["symptom_predicted_disease"] = predict_disease(symptoms_list)

    # Weighted average for overall health score
    weights = {
        "heart": 0.2,
        "diabetes": 0.15,
        "kidney": 0.15,
        "liver": 0.15,
        "stroke": 0.2
    }
    overall_score = sum(scores[d] * weights[d] for d in weights)

    return overall_score, scores
```
